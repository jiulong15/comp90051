```python
import pandas as pd

# 从CSV文件中导入数据并储存为DataFrame
df1 = pd.read_csv('train_data.csv')
df2 = pd.read_csv('train_data_merged.csv')
df3 = pd.read_csv('more_features.csv')
```

```python
df1 = df1.drop(['LOG_ID', 'PAT_ID', 'MRN'], axis=1)
df2 = df2.drop(['LOG_ID', 'PAT_ID', 'MRN'], axis=1)
df3 = df3.drop(['LOG_ID', 'PAT_ID', 'MRN'], axis=1)
# print(df1)
# print(df2)
print(df3)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler

# 假设 df1 是你的数据集
# 将数据集分为特征X和目标y
X = df1.drop('Cardiovascular', axis=1)  # 'target_column' 是你要预测的标签列
y = df1['Cardiovascular']

# 将数据集划分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 初始化标准化工具 (StandardScaler) 进行归一化
scaler = StandardScaler()

# 对训练集和测试集进行归一化
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 初始化SVM分类器
svm_model = SVC()

# 使用归一化后的数据训练模型
svm_model.fit(X_train_scaled, y_train)

# 使用模型进行预测
y_pred = svm_model.predict(X_test_scaled)

# 计算并显示准确率、精确率和召回率
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='binary')  # 'binary' 适用于二分类问题
recall = recall_score(y_test, y_pred, average='binary')

# 显示结果
print(f"准确率 (Accuracy): {accuracy}")
print(f"精确率 (Precision): {precision}")
print(f"召回率 (Recall): {recall}")

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler

# 将数据集分为特征X和目标y
X = df3.drop('Cardiovascular', axis=1)  # 'target_column' 是你要预测的标签列
y = df3['Cardiovascular']

# 将数据集划分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 初始化标准化工具 (StandardScaler) 进行归一化
scaler = StandardScaler()

# 对训练集和测试集进行归一化
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 初始化SVM分类器
svm_model = SVC()

# 使用归一化后的数据训练模型
svm_model.fit(X_train_scaled, y_train)

# 使用模型进行预测
y_pred = svm_model.predict(X_test_scaled)

# 计算并显示准确率、精确率和召回率
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='binary')  # 'binary' 适用于二分类问题
recall = recall_score(y_test, y_pred, average='binary')

# 显示结果
print(f"准确率 (Accuracy): {accuracy}")
print(f"精确率 (Precision): {precision}")
print(f"召回率 (Recall): {recall}")
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline



# 假设你已经有特征矩阵 X 和标签 y
# X 是你的输入特征，y 是目标变量
svm_model = make_pipeline(StandardScaler(), SVC(kernel='rbf'))
# 创建SVM模型实例

# 使用20折交叉验证
cv_scores = cross_val_score(svm_model, X, y, cv=20, scoring='precision')

# 输出每一折的交叉验证得分
print(f'20-fold Cross-Validation precision Scores: {cv_scores}')

# 输出平均准确率和标准差
print(f'Mean precision: {cv_scores.mean()}')
print(f'Standard Deviation: {cv_scores.std()}')

```

```python
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize

# 假设你已经有特征矩阵 X 和目标标签 y
# 将数据分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 创建SVM模型并启用概率估计
svm_model = SVC(kernel='rbf', probability=True)  # 确保使用 probability=True

# 训练模型
svm_model.fit(X_train, y_train)

# 预测概率
y_score = svm_model.predict_proba(X_test)[:, 1]  # 取正类的概率

# 计算ROC曲线
fpr, tpr, thresholds = roc_curve(y_test, y_score)

# 计算AUC
roc_auc = auc(fpr, tpr)

# 绘制ROC曲线
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif

# 假设 X 是一个 pandas DataFrame，y 是目标标签
# 将数据分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 选择最好的10个特征
selector = SelectKBest(f_classif, k=10)
X_train_new = selector.fit_transform(X_train, y_train)
X_test_new = selector.transform(X_test)

# 打印选中的特征名称
selected_feature_indices = selector.get_support(indices=True)
selected_feature_names = [X.columns[i] for i in selected_feature_indices]  # 确保 X 是一个 DataFrame
print(f"被选中的特征名称: {selected_feature_names}")

# 创建SVM模型并启用概率估计
svm_model = SVC(kernel='rbf', probability=True)

# 训练模型
svm_model.fit(X_train_new, y_train)

# 预测概率
y_score = svm_model.predict_proba(X_test_new)[:, 1]

# 计算ROC曲线
fpr, tpr, thresholds = roc_curve(y_test, y_score)

# 计算AUC
roc_auc = auc(fpr, tpr)

# 绘制ROC曲线
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
