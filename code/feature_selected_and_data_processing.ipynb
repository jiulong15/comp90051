```python
import pandas as pd
df_First_eight_features = pd.read_csv('First_eight_features.csv')
print(df_First_eight_features)
```

```python
import numpy as np
def height_to_inches(height):
    if pd.isna(height):
        return np.nan
    feet, inches = height.split("' ")
    return (int(feet) * 12 + float(inches))*2.54
df_eight_features_processing=df_First_eight_features.copy()
df_eight_features_processing['HEIGHT'] = df_eight_features_processing['HEIGHT'].apply(height_to_inches)

# caculate the mean
mean_height = df_eight_features_processing['HEIGHT'].mean()
print(mean_height)
# fulfill the NaNs
df_eight_features_processing['HEIGHT'].fillna(mean_height, inplace=True)

print(df_eight_features_processing)
```

```python
df_eight_features_processing['ICU_ADMIN_FLAG'] = df_eight_features_processing['ICU_ADMIN_FLAG'].replace({'Yes': 1, 'No': 0})
print(df_eight_features_processing)
```

```python
df_eight_features_processing['SEX'] = df_eight_features_processing['SEX'].replace({"Female": 0, "Male": 1})
print(df_eight_features_processing)
```

```python
abnormal_flag_mapping = {
    'N': 0,
    'L': 1,
    'H': 2,
    'LL': 3,
    'HH': 4
}

# replace the Abnormal Flag
df_eight_features_processing['Abnormal Flag'] = df_eight_features_processing['Abnormal Flag'].replace(abnormal_flag_mapping)

# caculate the mean of Abnormal Flag
mean_abnormal_flag = df_eight_features_processing['Abnormal Flag'].mean()

# use the mean to replace the 'NaN' in Abnormal Flag
df_eight_features_processing['Abnormal Flag'].fillna(mean_abnormal_flag, inplace=True)

print(df_eight_features_processing)
```

```python
df_eight_features_processing_selected=df_eight_features_processing[['Cardiovascular','LOS','ICU_ADMIN_FLAG','ICU_ADMIN_FLAG','HEIGHT','SEX','ASA_RATING_C','Abnormal Flag']]
df_eight_features_processing_selected = df_eight_features_processing_selected.apply(lambda x: x.fillna(x.mean()) if x.dtype.kind in 'biufc' else x)
# df_selected['column_name'] = df_selected['column_name'].fillna(df_selected['column_name'].mean())
print(df_eight_features_processing_selected)
```

```python
df_eight_features_processing= df_eight_features_processing.apply(lambda x: x.fillna(x.mean()) if x.dtype.kind in 'biufc' else x)
print(df_eight_features_processing)
df_eight_features_processing.to_csv('eight_features_processed.csv', index=False)
```
